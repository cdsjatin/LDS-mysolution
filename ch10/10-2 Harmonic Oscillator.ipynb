```python
# using sympy
s, w, t = symbols('s w t')
A = Matrix([[0,w],[-w,0]])
resolvent = Matrix(s*eye(2) - A)**-1
print(resolvent)

state_transistion_matrix = Matrix([[cos(w*t),sin(w*t)],[-sin(w*t),cos(w*t)]])
```

### (B)

```python
import matplotlib.pyplot as plt
```

```python
B = np.asarray(A.subs({w:1}))

print(B)
```

```python
# Grid of x, y points
nx, ny = 10, 10
x = np.linspace(-2, 2, nx)
y = np.linspace(-2, 2, ny)
X, Y = np.meshgrid(x, y)

U = np.zeros_like(X)
V = np.zeros_like(Y)

for i in range(nx):
    for j in range(ny):
        U[i,j],V[i,j] = B.dot(np.asarray([X[i,j],Y[i,j]]))

print(U.shape)

plt.quiver(X,Y,U,V)
plt.xlim(-2,2)
plt.ylim(-2,2)

plt.show()
```
